```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('multisim_dataset.parquet')
```

```python
df.set_index('telephone_number')
df = df.reset_index(drop=True)
```

```python
"""	trf  tariff used by customer
		age  customer age in years
		gndr  gender can be M  male, F  female, U  unknown
		tenure  customer tenure in days
		age_dev  age of device
		dev_man  device manufacturer
		device_os_name  OS on the device
		dev_num  number of devices that the sim card was installed in
		is_dualsim  devices capability to hold 2 sim cards simultaneously, 0  No, 1 - Yes
		is_featurephone  if device is featurephone, 0  No, 1 - Yes
		is_smartphone  if device is smartphone, 0  No, 1  Yes
		simcard_type  type of simcard (4G, 3G and etc)
		region  region of customer
		target  if subscriber is a multisimmer, 0 -No, 1  Yes"""
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
one_hot = pd.get_dummies(df['trf'], prefix='trf').astype(int)
df = df.join(one_hot)
df.drop(columns=['trf'], inplace=True)
```

```python
df.head()
```

```python
df['is_female'] = df['gndr'].apply(lambda x: 1 if x == 'F' else 0)
df.drop(columns=['gndr'], inplace=True)
```

```python
df['dev_man'].value_counts().plot(kind='bar')
plt.xlabel('Dev man')
plt.ylabel('Count')
plt.title('Count of Unique Values in Dev man')
plt.show()
```

```python
top10 = df['dev_man'].value_counts().nlargest(10)

print(top10)
```

```python
top5 = df['dev_man'].value_counts().nlargest(5).index.tolist()

df['dev_man_mod'] = df['dev_man'].apply(lambda x: x if x in top5 else 'unknown_dev')

one_hot = pd.get_dummies(df['dev_man_mod'], prefix='dev_man').astype(int)

df = df.join(one_hot)

df.drop(columns=['dev_man', 'dev_man_mod'], inplace=True)
```

```python
top3 = df['device_os_name'].value_counts().nlargest(3).index.tolist()

df['device_os_name_mod'] = df['device_os_name'].apply(lambda x: x if x in top3 else 'unknown_os_name')

one_hot = pd.get_dummies(df['device_os_name_mod'], prefix='device_os_name').astype(int)

df = df.join(one_hot)

df.drop(columns=['device_os_name', 'device_os_name_mod'], inplace=True)
```

```python
top3 = df['simcard_type'].value_counts().index.tolist()

df['simcard_type_mod'] = df['simcard_type'].apply(lambda x: x if x in top3 else 'simcard_type_unknown')

one_hot = pd.get_dummies(df['simcard_type_mod'], prefix='simcard_type').astype(int)

df = df.join(one_hot)

df.drop(columns=['simcard_type', 'simcard_type_mod'], inplace=True)
```

```python
top3 = df['region'].value_counts().nlargest(3).index.tolist()

df['region_mod'] = df['region'].apply(lambda x: x if x in top3 else 'other_regions')

one_hot = pd.get_dummies(df['region_mod'], prefix='region').astype(int)

df = df.join(one_hot)

df.drop(columns=['region', 'region_mod'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
float_cols = ['age', 'tenure', 'age_dev']
int_cols = [
    'dev_num', 'is_dualsim', 'is_featurephone', 'is_smartphone', 'trf_A', 'trf_B', 'trf_C',
    'trf_D', 'trf_E', 'trf_F', 'trf_G', 'trf_H', 'trf_I', 'trf_J', 'trf_K', 'trf_L', 'trf_M',
    'trf_N', 'trf_O', 'trf_P', 'trf_Q', 'trf_R', 'trf_T', 'is_female', 'dev_man_Apple',
    'dev_man_Hmd Global Oy', 'dev_man_Nokia', 'dev_man_Samsung', 'dev_man_Xiaomi',
    'dev_man_unknown_dev', 'device_os_name_Android OS', 'device_os_name_Proprietary',
    'device_os_name_iOS', 'device_os_name_unknown_os_name', 'simcard_type_3G',
    'simcard_type_4G', 'simcard_type_simcard_type_unknown', 'simcard_type_unknown',
    'region_BAKU', 'region_KHIRDALAN', 'region_SUMGAYIT', 'region_other_regions'
]

for col in float_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in int_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
```

```python
df.set_index('telephone_number')
df = df.reset_index(drop=True)

df = df.drop(columns=['telephone_number'])

y = df['is_dualsim'].rename('target')
x = df.drop(columns=['target', 'is_dualsim'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1)

param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],
    'subsample': [0.7, 1.0],
    'colsample_bytree': [0.7, 1.0]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=3, verbose=2, n_jobs=-1)

grid_search.fit(x_train, y_train)

print("Best parameters found: ", grid_search.best_params_)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(x_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)

plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score

y_pred = best_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f"Test Accuracy: {accuracy:.4f}")
print(f"Test Recall: {recall:.4f}")
print(f"Test Precision: {precision:.4f}")
```
